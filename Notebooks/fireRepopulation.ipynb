```python
df = pd.DataFrame(data)

```

```python
#not working again , and again, and again...

for col in df.columns[1:]:  # Skipping the first column as it contains County names
    df[col] = df[col].str.replace(',', '.').astype(float).fillna(0).astype(int)
```

```python
# new code to tranform to intergers. It is a new way, not learned before. 

def convert_to_integers(df):
    for col in df.columns[1:]:  # Skipping the first column as it contains County names
        df[col] = df[col].astype(str).str.replace(',', '.').astype(float).fillna(0).astype(int)
    return df

```

```python
# Apply the conversion
df = convert_to_integers(df)
```

```python

# Display the first few rows and the data types after conversion
print(df.head())
print(df.dtypes)
```

```python
# Sum the values of each column without the first column ('county')
column_sums = df.iloc[:, 1:].sum()  
```

```python
# Display the column sums
print(column_sums)
```

```python
# Create a new DataFrame from the summed values
sums_df = pd.DataFrame(column_sums, columns=['Total']).reset_index()
sums_df.columns = ['year', 'total']

# Display the new DataFrame
print(sums_df)
```

```python
# Save the new DataFrame to a CSV file
output_file_path = 'summed_totals_hafires.csv'  
sums_df.to_csv(output_file_path, index=False)

# Display the new DataFrame
print(sums_df)
```
