```python
df = pd.read_csv("fire_horainici_8523.csv")
df.head()
```

```python
import plotly.graph_objects as go
import numpy as np
```

```python
# more info of the plotly https://stackoverflow.com/questions/72595317/is-it-possible-to-generate-a-clock-chart-using-plotly
```

```python
# https://plotly.com/python/plotly-express/
```

```python
r = df['Average'].tolist()
```

```python
theta = np.arange(7.5,368,15)
width = [15]*24
```

```python
ticktexts = [f'$\large{i}$' if i % 6 == 0 else '' for i in np.arange(24)]
```

```python
import plotly.express as px
```

```python
fig = go.Figure(go.Barpolar(
    r=r,
    theta=theta,
    width=width,
    marker_color=df['Average'],
    marker_colorscale='Reds',
    marker_line_color="white",
    marker_line_width=2,
    opacity=0.8
))
```

```python

fig.update_layout(
    title = "Fires mostly start during the central part of the day",
    font=dict(
        family="Arial, monospace",
        size=15,
        color="Black"),
    template=None,
    polar=dict(
        hole=0.4,
        bgcolor='rgb(223, 223,223)',
        radialaxis=dict(
            showticklabels=False,
            ticks='',
            linewidth=2,
            linecolor='white',
            showgrid=False,
        ),
        angularaxis=dict(
            tickvals=np.arange(0,360,15),
            ticktext = ['0', '', '', '', '', '', '6', '', '', '', '', '', '12', '', '', '', '', '', '18', '', '', '', '', ''],
            showline=True,
            direction='clockwise',
            period=24,
            linecolor='white',
            gridcolor='white',
            showticklabels=True,
            ticks=''
        )
    )
)

fig.show()

```
