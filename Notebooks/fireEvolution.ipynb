```python
#changing the dateformat
import datetime
```

```python
from datetime import datetime
```

```python
oldformat = '11/01/2011'
```

```python
datetimeobject = datetime.strptime(oldformat,'%d/%m/%Y')
```

```python
datenewformat = datetimeobject.strftime('%m-%d-%Y')
```

```python
print(datenewformat)
```

```python
df2 = pd.DataFrame(df)
```

```python
df2['datenewformat'] = datenewformat
```
