## What is the proportion of forest repopulated regarding the burned surface?

```python
import numpy as np
```

```python
# First I have to change the NaN cells to a number. In that case, we will say there is no data

combined_df['total'] = combined_df['total'].replace({
            '0': np.nan
})

#this option doesn't do anything...so i decided to look for another...and I find the next above. 
```

```python
#Replace NaN in the 'total' column with 0
combined_df['total'] = combined_df['total'].fillna(0)
```

```python
combined_df
```

```python
#Now I can calculate the total of the burned and replanted are in roder to know the proportion. 
total_burned = combined_df['ha_forest'].sum()
total_burned
```

```python
total_repopulated = combined_df['total'].sum(),
total_repopulated
```

```python
(total_repopulated/total_burned) * 100
```
